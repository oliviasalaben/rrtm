```python
import csv

sza_val =[]
Rs_val = []
net_TOA_flux =[] 

# Open the CSV file
#with open('sza and Rs  - Sheet1.csv', 'r') as file:
with open('Copy of sza and Rs  - Sheet1.csv', 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Iterate over each row in the CSV file
    for row in csv_reader:
        # Each row is a list of values
        #print(row[1])
        sza_val.append(row[1])
        Rs_val.append(row[0])
        net_TOA_flux.append(row[4])
```

```python
sza_vals = np.array(sza_val[2:])
print(len(sza_vals))
no_quotes_sza = np.zeros(len(sza_vals))
for i in range(len(sza_vals)):
    no_quotes_sza[i] = sza_vals[i]
print(no_quotes_sza)

Rs_vals = np.array(Rs_val[2:])
print(len(Rs_vals))
no_quotes_Rs = np.zeros(len(Rs_vals))
for i in range(len(Rs_vals)):
    no_quotes_Rs[i] = Rs_vals[i]
print(no_quotes_Rs)

net_TOA_fluxes = np.array(net_TOA_flux[2:])
print(len(net_TOA_fluxes))
no_quotes_fluxes = np.zeros(len(net_TOA_fluxes))
for i in range(len(net_TOA_fluxes)):
    no_quotes_fluxes[i] = net_TOA_fluxes[i]
print(no_quotes_fluxes)
```

```python
# Example arrays of x coordinates, y coordinates, and values

# Plot the values at the specified coordinates
plt.scatter(no_quotes_sza, no_quotes_Rs, c=no_quotes_fluxes, cmap='viridis', s=100, edgecolors='k')  # s determines the size of points
plt.colorbar(label='Difference in Fluxes [W$m^-2$]')  # Add colorbar to show the values
plt.xlabel('Solar Zenith Angle')
plt.ylabel('Surface Albedo')
plt.title('Scatter Plot of Differences in Aerosol and Clear-sky fluxes at TOA')
plt.grid(True)
plt.show()
```

```python
import os

sza_Rs_pairs = []

# Path to the directory containing the files
folder_path = "rrtm_output"

# List all files in the directory
files = os.listdir(folder_path)

# Initialize dictionaries to store the paths of aerosol and clear files
aerosol_files = {}
clear_files = {}

# Filter files based on the pattern and read the contents
for file in files:
    if ".aerosol" in file:
        # Extract sza and Rs values from aerosol filename
        sza_start = file.find('sza') + 3
        sza_end = file.find('.aerosol')
        Rs_start = file.rfind('Rs') + 2
        Rs_end = file.rfind('_sza')
        sza = file[sza_start:sza_end]
        Rs = file[Rs_start:Rs_end]
        aerosol_files[(Rs, sza)] = os.path.join(folder_path, file)
        sza_Rs_pairs.append((Rs, sza))
    elif ".clear" in file:
        # Extract sza and Rs values from clear filename
        sza_start = file.find('sza') + 3
        sza_end = file.find('.clear')
        Rs_start = file.rfind('Rs') + 2
        Rs_end = file.rfind('_sza')
        sza = file[sza_start:sza_end]
        Rs = file[Rs_start:Rs_end]
        clear_files[(Rs, sza)] = os.path.join(folder_path, file)

# Pair aerosol and clear files based on matching sza and Rs values
for key in aerosol_files.keys():
    if key in clear_files:
        aerosol_path = aerosol_files[key]
        clear_path = clear_files[key]
        Rs, sza = key
        Kevin_read_output_rrtm.read_output_rrtm(clear_path,aerosol_path)
        #read_output_rrtm.read_OUTPUT_RRTM(clear_path,aerosol_path)
        
```

```python
def process_clear_and_aerosol_files(folder_path):
    clear_files = []
    aerosol_files = []
    
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Filter files with .clear and .aerosol extensions
    for file in files:
        if file.endswith(".clear"):
            clear_files.append(file)
        elif file.endswith(".aerosol"):
            aerosol_files.append(file)
    
    return clear_files, aerosol_files
            

# Path to the directory containing the files
folder_path = "rrtm_output"
process_clear_and_aerosol_files(folder_path)
```
