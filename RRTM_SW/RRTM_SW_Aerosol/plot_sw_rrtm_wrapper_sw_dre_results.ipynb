## Download North African dust optical properties 0W, 18N results 

```python
#Download North African SW RRTM Wrapper output
NAf_0W_18N_TOA = np.loadtxt('NAf_0W_18N_TOA_file_shrink.txt', skiprows=1)
NAf_0W_18N_seas = []
NAf_0W_18N_lat = []
NAf_0W_18N_lon = []
NAf_0W_18N_Rs = []
NAf_0W_18N_sza = []
NAf_0W_18N_aerosol = []
NAf_0W_18N_clear = []
NAf_0W_18N_net = []
NAf_0W_18N_net_daod = 0.622

for i in range(len(NAf_0W_18N_TOA)):
    NAf_0W_18N_seas.append(NAf_0W_18N_TOA[i][0])
    NAf_0W_18N_lat.append(NAf_0W_18N_TOA[i][1])
    NAf_0W_18N_lon.append(NAf_0W_18N_TOA[i][2])
    NAf_0W_18N_Rs.append(NAf_0W_18N_TOA[i][3])
    NAf_0W_18N_sza.append(NAf_0W_18N_TOA[i][4])
    NAf_0W_18N_aerosol.append(NAf_0W_18N_TOA[i][5])
    NAf_0W_18N_clear.append(NAf_0W_18N_TOA[i][6])
    NAf_0W_18N_net.append(NAf_0W_18N_TOA[i][7])
```

```python
NAf_0W_18N_seas = np.array(NAf_0W_18N_seas)
NAf_0W_18N_lat = np.array(NAf_0W_18N_lat)
NAf_0W_18N_lon = np.array(NAf_0W_18N_lon)
NAf_0W_18N_Rs = np.array(NAf_0W_18N_Rs)
NAf_0W_18N_sza = np.array(NAf_0W_18N_sza)
NAf_0W_18N_aerosol = np.array(NAf_0W_18N_aerosol)
NAf_0W_18N_clear = np.array(NAf_0W_18N_clear)
NAf_0W_18N_net = np.array(NAf_0W_18N_net)
```

```python
NAf_SWDRE_array = np.zeros((4,len(np.unique(NAf_0W_18N_lat)),len(np.unique(NAf_0W_18N_lon))))
lat_coord = np.unique(NAf_0W_18N_lat)                           
lon_coord = np.unique(NAf_0W_18N_lon)

for i in range(len(NAf_0W_18N_TOA)):
    for iseas in range(0,4):
        for ilat in range(len(np.unique(NAf_0W_18N_lat))):
            for ilon in range(len(np.unique(NAf_0W_18N_lon))):
                if iseas == NAf_0W_18N_seas[i] and lat_coord[ilat] == NAf_0W_18N_lat[i]and lon_coord[ilon] == NAf_0W_18N_lon[i]:                            
                    NAf_SWDRE_array[iseas,ilat,ilon] = NAf_0W_18N_net[i] #/2 #divide by 2 because sza use is only for daylight hours, and we need full day (including nighttime)
```

## Multiply North African DRE by DustCOMM DAOD --dust optical properties 0W, 18N

```python
daod_Dustcomm_PM20_Dataset = Dataset('../../Dust_Optical_Properties/DustCOMM_source_region_DAOD_seas_PM20_abs.nc')
print(daod_Dustcomm_PM20_Dataset) #[lat,lon,bin,source,season]

lat_Dustcomm_PM20 = np.array(daod_Dustcomm_PM20_Dataset.variables['lat'][:])
lon_Dustcomm_PM20 = np.array(daod_Dustcomm_PM20_Dataset.variables['lon'][:])
mean_Daod_Dustcomm_20PM = np.array(daod_Dustcomm_PM20_Dataset.variables['Mean'][:])
#dustcomm_seas = np.array(daod_Dustcomm_PM20_Dataset.variables['season'][:])
print(mean_Daod_Dustcomm_20PM.shape)

sum_North_African_Mean_Daod_20PM = (mean_Daod_Dustcomm_20PM[:,:,0,:] +mean_Daod_Dustcomm_20PM[:,:,1,:] +mean_Daod_Dustcomm_20PM[:,:,2,:])
print(sum_North_African_Mean_Daod_20PM.shape)

bulk_NAf_daod = np.zeros((96,144,4))
for ilat in range(len(lat_Dustcomm_PM20)):
    for ilon in range(len(lon_Dustcomm_PM20)):
        for iseas in range(4):
            bulk_NAf_daod[ilat,ilon,iseas] = np.nansum(sum_North_African_Mean_Daod_20PM[ilat,ilon,iseas])
print(bulk_NAf_daod.shape)
```

```python
def get_month_albedo(year, month):
    filename = f'MERRA2_400.tavgM_2d_rad_Nx.{year}{month:02}.nc4'
    dataset = Dataset(filename)
    albedo = np.array(dataset.variables['ALBEDO'][0])
    return albedo

albedo_2023 = []

for month in range(1, 13):
    albedo_2023.append(get_month_albedo(2023, month))

albedo_2023 = np.array(albedo_2023)
```
