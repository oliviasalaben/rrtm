```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python

# Example arrays of x coordinates, y coordinates, and values
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
values = [5, 10, 15, 20, 25]

# Plot the values at the specified coordinates
plt.scatter(x, y, c=values, cmap='viridis', s=100, edgecolors='k')  # s determines the size of points
plt.colorbar(label='Values')  # Add colorbar to show the values
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Values at Coordinates')
plt.grid(True)
plt.show()
```

```python
import csv

sza_val =[]
Rs_val = []
net_TOA_flux =[] 

# Open the CSV file
with open('sza and Rs  - Sheet1.csv', 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Iterate over each row in the CSV file
    for row in csv_reader:
        # Each row is a list of values
        #print(row[1])
        sza_val.append(row[1])
        Rs_val.append(row[0])
        net_TOA_flux.append(row[4])
```

```python
sza_vals = np.array(sza_val[2:])
print(sza_vals)
no_quotes_sza = np.zeros(20)
for i in range(len(sza_vals)):
    no_quotes_sza[i] = sza_vals[i]
print(no_quotes_sza)

Rs_vals = np.array(Rs_val[2:])
print(Rs_vals)
no_quotes_Rs = np.zeros(20)
for i in range(len(Rs_vals)):
    no_quotes_Rs[i] = Rs_vals[i]
print(no_quotes_Rs)

net_TOA_fluxes = np.array(net_TOA_flux[2:])
print(net_TOA_fluxes)
no_quotes_fluxes = np.zeros(20)
for i in range(len(net_TOA_fluxes)):
    no_quotes_fluxes[i] = net_TOA_fluxes[i]
print(no_quotes_fluxes)
```

```python
# Example arrays of x coordinates, y coordinates, and values

# Plot the values at the specified coordinates
plt.scatter(no_quotes_sza, no_quotes_Rs, c=no_quotes_fluxes, cmap='viridis', s=100, edgecolors='k')  # s determines the size of points
plt.colorbar(label='Difference in Fluxes [W$m^-2$]')  # Add colorbar to show the values
plt.xlabel('Solar Zenith Angle')
plt.ylabel('Surface Albedo')
plt.title('Scatter Plot of Differences in Aerosol and Clear-sky fluxes at TOA')
plt.grid(True)
plt.show()
```

```python
plt.plot(no_quotes_sza[0:5],no_quotes_fluxes[0:5],label='Rs=0.4')
plt.plot(no_quotes_sza[5:10],no_quotes_fluxes[5:10],label='Rs=0.0.37')
plt.plot(no_quotes_sza[10:15],no_quotes_fluxes[10:15],label='Rs=0.33')
plt.plot(no_quotes_sza[15:],no_quotes_fluxes[15:],label='Rs=0.30')
# Add horizontal line at zero
plt.axhline(0, color='k', linestyle='--')
plt.legend()
plt.title('Clear-sky minus Aerosol fluxes at TOA as function of surface albedo')
plt.xlabel('Solar Zenith in Degrees [˚]')
plt.ylabel('Difference in Fluxes [W$m^-2$]')
```
