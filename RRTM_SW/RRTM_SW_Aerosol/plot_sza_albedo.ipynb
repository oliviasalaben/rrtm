## North African dust optical properties 0W, 18N results 

```python
import csv

sza_val =[]
Rs_val = []
net_TOA_flux =[] 

# Open the CSV file
#with open('sza and Rs  - Sheet1.csv', 'r') as file:
with open('Copy of sza and Rs  - Sheet1.csv', 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Iterate over each row in the CSV file
    for row in csv_reader:
        # Each row is a list of values
        #print(row[1])
        sza_val.append(row[1])
        Rs_val.append(row[0])
        net_TOA_flux.append(row[4])
```

```python
sza_vals = np.array(sza_val[2:])
print(len(sza_vals))
no_quotes_sza = np.zeros(len(sza_vals))
for i in range(len(sza_vals)):
    no_quotes_sza[i] = sza_vals[i]
print(no_quotes_sza)

Rs_vals = np.array(Rs_val[2:])
print(len(Rs_vals))
no_quotes_Rs = np.zeros(len(Rs_vals))
for i in range(len(Rs_vals)):
    no_quotes_Rs[i] = Rs_vals[i]
print(no_quotes_Rs)

net_TOA_fluxes = np.array(net_TOA_flux[2:])
print(len(net_TOA_fluxes))
no_quotes_fluxes = np.zeros(len(net_TOA_fluxes))
for i in range(len(net_TOA_fluxes)):
    no_quotes_fluxes[i] = net_TOA_fluxes[i]
print(no_quotes_fluxes)
```

```python
# Example arrays of x coordinates, y coordinates, and values

# Plot the values at the specified coordinates
plt.scatter(no_quotes_sza, no_quotes_Rs, c=no_quotes_fluxes, cmap='viridis', s=100, edgecolors='k')  # s determines the size of points
plt.colorbar(label='Difference in Fluxes [W$m^-2$]')  # Add colorbar to show the values
plt.xlabel('Solar Zenith Angle')
plt.ylabel('Surface Albedo')
plt.title('Scatter Plot of Differences in Aerosol and Clear-sky fluxes at TOA')
plt.grid(True)
plt.show()
```
